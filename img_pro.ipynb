```python
w, h = im2.width, im2.height
img =im2.resize((w//2, h//2)) # reducing size by 50%
img
```

```python
# enhancing color
en = ImageEnhance.Color(img)
```

```python
for i in range(-5, 6):
    en.enhance(i).save(f"img/img_enhance_{i}.png")
```

```python
# image drawing
from PIL import ImageFont
img =im2.resize((w//2, h//2))
print(img.size)
dr = ImageDraw.Draw(img)
start_point = (0,img.height//2)
end_point = (img.width, img.height//2)
dr.line(xy=(start_point, end_point), fill='white', width=5)
dr.line(xy=((img.width//2, 0),(img.width//2, img.height)),fill='white', width=5)
dr.circle((250, 175), radius=50, fill='red')
dr.circle((250, 175), radius=100, outline='yellow', width=40)
font = ImageFont.truetype(r"c:\WINDOWS\Fonts\GOODTIME.TTF", 50)
dr.text((250,175), "Off Target", fill='black', font=font, 
        stroke_fill='white', stroke_width=5, anchor='ms')
img.save('img/generated.png')
img
```
